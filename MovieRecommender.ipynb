# Movie Recommender

```python
train_data,test_data = movie_data.random_split(.8,seed=0)
```

## Popularity recommender without target

```python
popularity_model1 = gl.popularity_recommender.create(train_data,
                                                         user_id="userId",
                                                         item_id="movieId")
```

### Usando el modelo

```python
popularity_model1.recommend(users=[usuarios[0]])
```

```python
popularity_model1.recommend(users=[usuarios[1]])
```

## Popularity recommender with target

```python
popularity_model2 = gl.popularity_recommender.create(train_data,
                                                         user_id="userId",
                                                         item_id="movieId",
                                                         target="rating")
```

### Usando el modelo

```python
popularity_model2.recommend(users=[usuarios[0]])
```

```python
popularity_model2.recommend(users=[usuarios[1]])
```

## Similarity recommender without target

```python
similarity_model1 = gl.item_similarity_recommender.create(train_data,
                                                                user_id='userId',
                                                                item_id='movieId')
```

```python
similarity_model1.recommend(users=[usuarios[0]])
```

```python
similarity_model1.recommend(users=[usuarios[1]])
```

## Similarity recommender with target

```python
similarity_model2 = gl.item_similarity_recommender.create(train_data,
                                                                user_id='userId',
                                                                item_id='movieId',
                                                                target='rating')
```

```python
similarity_model2.recommend(users=[usuarios[0]])
```

```python
similarity_model2.recommend(users=[usuarios[1]])
```

## Ranking factorizacion model without target

```python
factorization_model1 = gl.ranking_factorization_recommender.create(train_data,
                                                                        user_id='userId',
                                                                        item_id='movieId',
                                                                  max_iterations=500)
```

```python
factorization_model1.recommend(users=[usuarios[0]])
```

```python
factorization_model1.recommend(users=[usuarios[15]])
```

## Ranking factorization model with target

```python
factorization_model2 = gl.ranking_factorization_recommender.create(train_data,
                                                                        user_id='userId',
                                                                        item_id='movieId',
                                                                       item_data=movies,
                                                                        target='rating',
                                                                        max_iterations=500)
```

```python
factorization_model2.recommend(users=[usuarios[0]])
```

```python
factorization_model2.recommend(users=[usuarios[1]])
```

## Ranking factorization model with target and max_iterations = 10,000

```python
factorization_model = gl.ranking_factorization_recommender.create(train_data,
                                                                        user_id='userId',
                                                                        item_id='movieId',
                                                                        target='rating',
                                                                        max_iterations=10000)
```

```python
factorization_model.recommend(users=[usuarios[0]])
```

```python
factorization_model.recommend(users=[usuarios[1]])
```

## Comparando los modelos

```python
model_performance = gl.compare(test_data, [popularity_model1, popularity_model2, similarity_model1, similarity_model2, factorization_model1, factorization_model2, factorization_model], user_sample=0.05)
gl.show_comparison(model_performance,[popularity_model1, popularity_model2, similarity_model1, similarity_model2, factorization_model1, factorization_model2, factorization_model])
```

```python
gl.canvas.set_target('browser')
```
